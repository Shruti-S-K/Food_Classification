```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
from keras.preprocessing import image
from keras.models import load_model
```

```python
model = load_model('vgg16_model.h5')
label = ['steak', 'cookies', 'burgers', 'hotdogs', 'toast', 'pizza','donuts', 'frenchfries', 'friedchicken', 'friedegg']
```

```python
def get_img(img_path):
  img = image.load_img(img_path,target_size=(224,224))
  img_arr = image.img_to_array(img)
  img_arr = np.expand_dims(img_arr,axis=0)
  img_arr = img_arr/255
  return img_arr
```

```python
image_path = '/content/drive/MyDrive/CNN Practice Datasets/Food Classification/Testing dataset/chicken_14.jpg'
new_img = get_img(image_path)
new_img
```

```python
pred = model.predict(new_img)
pred
```

```python
classid = np.argmax(pred, axis = 1)
classid
```

```python
label[int(classid)]
```

```python
plt.imshow(new_img[0])
plt.axis('off')
plt.show()
```

```python
img_path = '/content/drive/MyDrive/CNN Practice Datasets/Food Classification/Testing dataset/pizza_2.jpg'
new_img = get_img(img_path)
pred = model.predict(new_img)
class_id = np.argmax(pred,axis=1)
print(label[int(class_id)])
plt.imshow(new_img[0])
plt.axis('off')
plt.show()
```
